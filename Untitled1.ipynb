```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import pandas as pd

# Inicjalizacja trybu notebook
init_notebook_mode(connected=True)

# Przejście w tryb offline dla cufflinks
cf.go_offline()

# Wczytanie danych
df = pd.read_csv('OECDBLI2017.csv')

# Sortowanie i konwersja na DataFrame
df_sorted = df.set_index('Country')['Employment rate as pct'].sort_values(ascending=False).to_frame()

# Wywołanie iplot na DataFrame z określonymi kolorami

df_sorted.iplot(kind='bar', title='Employment Rate by Country', xTitle='Country', yTitle='Employment Rate (%)', color='#1f77b4')
```

```python
# Sprawdzenie nagłówków kolumn
print(df.columns)

# Sprawdzenie przykładowych danych
print(df.head())

# Sprawdzenie typów danych
print(df.dtypes)

# Sprawdzenie brakujących wartości
print(df.isnull().sum())
```

```python
# Ustawienie indeksu
df.set_index('Country', inplace=True)

# Sprawdzenie przykładowych danych po ustawieniu indeksu
print(df.head())

# Sortowanie wartości
df_sorted = df['Employment rate as pct'].sort_values(ascending=False)

# Sprawdzenie przykładowych danych po sortowaniu
print(df_sorted.head())

# Konwersja na DataFrame
df_sorted = df_sorted.to_frame()

# Sprawdzenie przykładowych danych po konwersji na DataFrame
print(df_sorted.head())

# Wywołanie iplot na DataFrame
df_sorted.iplot(kind='bar', title='Employment Rate by Country', xTitle='Country', yTitle='Employment Rate (%)', colors=['blue'])
```
